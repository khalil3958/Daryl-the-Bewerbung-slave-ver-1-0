```python
from lxml import html
import requests
import pandas 
import pandas as pd 
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError

pokedex=requests.get('https://pokemondb.net/pokedex/national')
pokedex_infos=html.fromstring(pokedex.content) 
names_table=pokedex_infos.xpath('//a[@class="ent-name"]/text()')
#dict(map(reversed, enumerate(names_table)))

def exceptions(name):
    exceptions=["Nidoran♀","Nidoran♂","Farfetch'd","Mr. Mime","Mime Jr.","Flabébé","Type: Null","Tapu Koko","Tapu Lele","Tapu Bulu","Tapu Fini"]
    if name==exceptions[0]:
        name="nidoran-f"
    elif name==exceptions[1]:
        name="nidoran-m"
    elif name==exceptions[2]:
        name="farfetchd"
    elif name==exceptions[3]:
        name="mr-mime"
    elif name==exceptions[4]:
        name="mime-jr"
    elif name==exceptions[5]:
        name="flabebe"
    elif name==exceptions[6]:
        name="type-null"
    elif name==exceptions[7]:
        name="tapu-koko"
    elif name==exceptions[8]:
        name="tapu-lele"
    elif name==exceptions[9]:
        name="tapu-bulu"
    elif name==exceptions[10]:
        name="tapu-fini"
    return name

def get_infos(url,path):
    real_list=[]
    
    for i in range (len(names_table)):
        inner_infos=requests.get(url+exceptions(str(names_table[i])))
        print ("loading...",(int((i/810)*100)),"%")
        inner_infos_text=html.fromstring(inner_infos.text)
        prep_list=inner_infos_text.xpath(path)
        if  (len(prep_list)>1):
            if (prep_list[1]==prep_list[0]) :
                   real_list.append(prep_list[0])
            else:
                myString = " ".join(prep_list )
                real_list.append(myString)
        else:
                myString = " ".join(prep_list )
                real_list.append(myString)
    return real_list
import numpy as np 
pokemon_infos=requests.get('https://pokemondb.net/pokedex/'+str(names_table[0]))
pokemon_infos1=html.fromstring(pokemon_infos.content)
dict_header=pokemon_infos1.xpath('//div[@class="grid-col span-md-6 span-lg-4"]//th/text()')
pok_num= get_infos('https://pokemondb.net/pokedex/','//div[@class="grid-col span-md-6 span-lg-4"]//td//strong/text()')  
pok_type= get_infos('https://pokemondb.net/pokedex/','//div[@class="grid-col span-md-6 span-lg-4"]//td//a[@class]/text()' )
pok_species= get_infos('https://pokemondb.net/pokedex/','//div[@class="grid-col span-md-6 span-lg-4"]//tr[3]//td/text()' )
pok_height= get_infos('https://pokemondb.net/pokedex/','//div[@class="grid-col span-md-6 span-lg-4"]//tr[4]//td/text()')
pok_weight= get_infos('https://pokemondb.net/pokedex/','//div[@class="grid-col span-md-6 span-lg-4"]//tr[5]//td/text()' )
pok_abilities= get_infos('https://pokemondb.net/pokedex/','//div[@class="grid-col span-md-6 span-lg-4"]//tr[6]//td//a/text()')
pok_local= get_infos('https://pokemondb.net/pokedex/','//div[@class="grid-col span-md-6 span-lg-4"]//tr[7]//td/text()')

#diccionnary={}
pokemon_data=pd.DataFrame(np.column_stack([pok_num,pok_type,pok_species,pok_height,pok_weight,pok_abilities,pok_local]), 
                               columns=dict_header , index=names_table)
#pokemon_data.from_records(, columns=)
#pokemon_data.index(names_table)
print(pokemon_data)
pokemon_data.to_csv(r'C:\Users\hp\Downloads\export_dataframe.csv')
```
