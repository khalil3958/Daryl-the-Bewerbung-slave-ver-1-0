```python
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError
```

```python
try:
    html = urlopen("https://www.facebook.com/DonaldTrump/?ref=br_rs")
except HTTPError as e:
    print("HTTP error")
except URLError as e:
    print("Server not found!")
else:
    c=html.read()
    print(c)
```

```python
try:
    html = urlopen("http://www.example.com/")
except HTTPError as e:
    print("HTTP error")
except URLError as e:
    print("Server not found!")
else:
    print("HTML Details")    
    print(html.read())  
```

```python
import requests
response = requests.get("https://en.wikipedia.org/robots.txt")
test = response.text  #response will be transefered to a text 
print("robots.txt for http://www.wikipedia.org/")
print("===================================================")
print(test)
```

```python
import requests #another way to check out i a certain page exists 
request = requests.get('http://www.example.com')
if request.status_code == 200:
    print('Web site exists')
else:
    print('Web site does not exist') 
```

```python
from lxml import html
import requests
response = requests.get('http://www.data.gov/')
c=response.text
c
```

```python
doc_gov = html.fromstring(response.content) #now contains the whole HTML file in a nice tree structure 
link_gov = doc_gov.cssselect('small a')[0]
print("Number of datasets currently listed on data.gov:")
print(link_gov.text)
```
